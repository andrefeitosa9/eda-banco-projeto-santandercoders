```python
def calcular_limites_outlier(dataframe, coluna):
    dados = dataframe[coluna]
    
    q1 = np.quantile(dados, 0.25)
    q3 = np.quantile(dados, 0.75)
    
    iqr = q3 - q1
    
    limite_inferior = q1 - 1.5*iqr
    limite_superior =  q3 + 1.5*iqr
    
    return limite_inferior, limite_superior

def eh_outlier(dado, dataframe, coluna):

    limite_inferior, limite_superior = calcular_limites_outlier(dataframe, coluna)
    if dado < limite_inferior or dado > limite_superior:
        return 1
    else: 
        return 0
```

```python
def calcular_limites_outlier_faixa_etaria(dataframe, coluna, faixa_etaria):

    dados_faixa_etaria = dataframe[dataframe['faixa_etaria'] == faixa_etaria]
    dados = dados_faixa_etaria[coluna]
    
    q1 = np.quantile(dados, 0.25)
    q3 = np.quantile(dados, 0.75)
    
    iqr = q3 - q1
    
    limite_inferior = q1 - 1.5*iqr
    limite_superior =  q3 + 1.5*iqr
    
    return limite_inferior, limite_superior


def eh_outlier_na_faixa_etaria(dado, dataframe, coluna, faixa_etaria):

    limite_inferior, limite_superior = calcular_limites_outlier_faixa_etaria(dataframe, coluna, faixa_etaria)
    if dado < limite_inferior or dado > limite_superior:
        return 1
    else: 
        return 0
```

```python
def calcular_limites_outlier_pais(dataframe, coluna, pais):

    dados_faixa_etaria = dataframe[dataframe['pais'] == pais]
    dados = dados_faixa_etaria[coluna]
    
    q1 = np.quantile(dados, 0.25)
    q3 = np.quantile(dados, 0.75)
    
    iqr = q3 - q1
    
    limite_inferior = q1 - 1.5*iqr
    limite_superior =  q3 + 1.5*iqr
    
    return limite_inferior, limite_superior


def eh_outlier_pais(dado, dataframe, coluna, pais):

    limite_inferior, limite_superior = calcular_limites_outlier_pais(dataframe, coluna, pais)
    if dado < limite_inferior or dado > limite_superior:
        return 1
    else: 
        return 0
```

```python
def fazer_graficos_histogramas(coluna, dataframe):
    print(f'\nPara a coluna {coluna}: \n')
    
    if coluna == 'faixa_etaria':
        plt.figure(figsize=(12, 8))
        dataframe[coluna] = dataframe[coluna].astype(str)
    
    # Se houver mais de 5 valores √∫nicos, exibe o describe
    if len(dataframe[coluna].unique().tolist()) > 5:
        print(dataframe[coluna].describe())
    
    # Se houver menos de 10 valores √∫nicos, exibe as contagens relativas
    if len(dataframe[coluna].unique().tolist()) < 10:
        print(dataframe[coluna].value_counts(1))
    print()

    # Boxplot e histograma combinados para colunas num√©ricas
    if coluna in colunas_com_boxplot:

        # Gr√°fico combinado de histograma e boxplot
        fig, (ax_box, ax_hist) = plt.subplots(
            nrows=2, 
            sharex=True, 
            gridspec_kw={"height_ratios": (0.25, 0.75)},
            figsize=(12, 8)
        )
        
        sns.boxplot(
            data=dataframe, x=coluna, ax=ax_box, showmeans=True, color="lightblue"
        )
        ax_box.set_title(f"Boxplot de {coluna}")

        # Histograma com barras empilhadas para 'saiu_banco'
        sns.histplot(
            data=dataframe,
            x=coluna,
            hue='saiu_banco',
            multiple='stack',
            palette={0: "blue", 1: "red"},
            ax=ax_hist
        )
        ax_hist.set_title(f"Histograma de {coluna}")
        ax_hist.set_xlabel(f"{coluna}")
        
        # Adiciona linhas de m√©dia e mediana no histograma
        media = dataframe[coluna].mean()
        mediana = dataframe[coluna].median()
        
        ax_hist.axvline(media, color="green", linestyle="--", linewidth=2)
        ax_hist.axvline(mediana, color="black", linestyle="-", linewidth=2)

        # Anota√ß√µes para m√©dia e mediana
        ax_hist.annotate(f'M√©dia: {media:.2f}', 
                         xy=(media, ax_hist.get_ylim()[1] * 0.85), 
                         xytext=(media + 1, ax_hist.get_ylim()[1] * 0.9),
                         textcoords='data',
                         color='green', fontsize=12,
                         bbox=dict(facecolor='white', alpha=0.6),
                         arrowprops=dict(facecolor='green', arrowstyle='->'))

        ax_hist.annotate(f'Mediana: {mediana:.2f}', 
                         xy=(mediana, ax_hist.get_ylim()[1] * 0.6), 
                         xytext=(mediana - 2, ax_hist.get_ylim()[1] * 0.65),
                         textcoords='data',
                         color='black', fontsize=12,
                         bbox=dict(facecolor='white', alpha=0.6),
                         arrowprops=dict(facecolor='black', arrowstyle='->'))

        plt.tight_layout()
        plt.show()

    # Apenas histograma para colunas categ√≥ricas
    else:
        plt.figure(figsize=(12, 8))
        sns.histplot(
            data=dataframe,
            x=coluna,
            hue='saiu_banco',
            multiple='stack',
            palette={0: "blue", 1: "red"}
        )
        
        plt.title(f"Histograma de {coluna}")
        plt.xlabel(f"{coluna}")
        plt.tight_layout()
        plt.show()

    # Propor√ß√µes e c√°lculos de impacto em sa√≠da ou n√£o
    nome_coluna = coluna + "_intervalada"
    df_proporcoes_corrigido = dataframe.copy()

    if len(dataframe[coluna].unique().tolist()) > 20:
        df_proporcoes_corrigido[nome_coluna] = pd.cut(dataframe[coluna], bins=20, right=True)
        df_proporcoes_corrigido[nome_coluna] = df_proporcoes_corrigido[nome_coluna].astype(str)

        df_proporcoes_agrupado = df_proporcoes_corrigido.groupby(nome_coluna)['saiu_banco'].value_counts(normalize=True).unstack()
        df_proporcoes_agrupado = df_proporcoes_agrupado.fillna(0)
        df_proporcoes_agrupado = df_proporcoes_agrupado * 100

        # Formata os valores como porcentagem com 2 casas decimais
        df_proporcoes_agrupado = df_proporcoes_agrupado.map(lambda x: f'{x:.2f}%')

        print(df_proporcoes_agrupado)

    else:
        df_proporcoes_agrupado = dataframe.groupby(coluna)['saiu_banco'].value_counts(normalize=True).unstack()
        df_proporcoes_agrupado = df_proporcoes_agrupado * 100

        # Formata os valores como porcentagem com 2 casas decimais
        df_proporcoes_agrupado = df_proporcoes_agrupado.map(lambda x: f'{x:.2f}%')

        print(df_proporcoes_agrupado)

    print("*-"*50)
    print()
```
